### Understanding LangChain Basics

```python
import os

from dotenv import load_dotenv
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
```

```python
load_dotenv()

model_name="llama3.2"
```

```python
if not model_name:
    raise ValueError("Please set the MODEL_NAME environment variable.")
```

```python
llm = ChatOllama(
    model=model_name,
    temperature=0.9,
    max_tokens=1000,
    base_url="http://localhost:11434"
)
```

```python
prompt_template = PromptTemplate(
    input_variables=["topic1", "topic2"],
    template="""
        You are a helpful assistant. 
        Your task is to generate a tweet that summarizes two topics.
        
        I want you to generate a tweet that summarizes the following input:
        
        {topic1}
        
        and 
        
        {topic2}
    """
)
```

```python
prompt = prompt_template.format(
    topic1="MIQ",
    topic2="Digital Transformation"
)

response = llm.invoke(prompt)

print(response.content)
```
